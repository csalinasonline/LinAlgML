```python
# matrix Hadamard product
from numpy import array
# define first matrix
A = array([
    [1, 2, 3],
    [4, 5, 6]
])
print(A)
#define second matrix
B = array([
    [2, 2, 2],
    [2, 2, 2]
])
print(B)
# multiply matrices
C = A * B
print(C)
```

```python
# matrix division
#define first matrix
A = array([
    [1, 2, 3],
    [4, 5, 6]
])
print(A)
#define second matrix
B = array([
    [2, 2, 2],
    [2, 2, 2]
])
print(B)
# divide matrices
C = A / B
print(C)
```
